## Preprocesar Datos

### Analizar Label

Analizar cuales son los posibles valores del label.

```python
census['income_bracket'].unique()
```

### Modificar el Objetivo (label)

Sustituir posibles valores del label por 0 y 1

```python
def label_fix(label):
    return 0 if label == " <=50K" else 1
```

```python
census['income_bracket'] = census['income_bracket'].apply(label_fix)
```

```python
census
```

### Separar los conjuntos de datos y obtener los datos de entrenamiento y test

```python
from sklearn.model_selection import train_test_split
```

```python
x_data = census.drop('income_bracket', axis=1)
```

```python
y_labels = census['income_bracket']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, y_labels, test_size=0.3, random_state=42)
```

# Create feature_columns for categorical features

### Vocabulary list

```python
gender = tf.feature_column.categorical_column_with_vocabulary_list('gender', ['Female', 'Male'])
```

### Hash bucket

```python
occupation = tf.feature_column.categorical_column_with_hash_bucket('occupation', hash_bucket_size=1000)
marital_status = tf.feature_column.categorical_column_with_hash_bucket('marital_status', hash_bucket_size=1000)
relationship = tf.feature_column.categorical_column_with_hash_bucket('relationship', hash_bucket_size=1000)
education = tf.feature_column.categorical_column_with_hash_bucket('education', hash_bucket_size=1000)
workclass = tf.feature_column.categorical_column_with_hash_bucket('workclass', hash_bucket_size=1000)
native_country = tf.feature_column.categorical_column_with_hash_bucket('native_country', hash_bucket_size=1000)
```

## Create feature_columns for the numeric features

```python
age = tf.feature_column.numeric_column('age')
education_num = tf.feature_column.numeric_column('education_num')
capital_gain = tf.feature_column.numeric_column('capital_gain')
capital_loss = tf.feature_column.numeric_column('capital_loss')
hours_per_week = tf.feature_column.numeric_column('hours_per_week')
```

```python
feat_cols = [gender, occupation, relationship, education, workclass, native_country, age, education_num, capital_gain, capital_loss, hours_per_week]
```

```python
feat_cols
```

# Part 3 - Model Selection

## Input Function

```python
def train_input_fn(features, labels, batch_size=100, num_epochs=None, shuffle=True):
    input_function = tf.estimator.inputs.pandas_input_fn(x=features, y=labels, batch_size=batch_size, num_epochs=num_epochs, shuffle=shuffle)
    return input_function
```

```python
# input_func = tf.estimator.inputs.pandas_input_fn(x=X_train, y=y_train, batch_size=100, num_epochs=None, shuffle=True)
```

## Clasificador Lineal (LinearClassifier)

```python
model = tf.estimator.LinearClassifier(feature_columns=feat_cols)
```

## Entrenar el Modelo

```python
model.train(input_fn=train_input_fn(X_train, y_train), steps=5000)
```

## Evaluar el Modelo

```python
def eval_input_fn(features, labels, batch_size, shuffle=False):
    eval_input = tf.estimator.inputs.pandas_input_fn(x=features, y=labels, batch_size=batch_size, shuffle=False)
    return eval_input
```

```python
# pred_fn = tf.estimator.inputs.pandas_input_fn(x=X_test, batch_size=len(X_test), shuffle=False)
```

```python
eval_result = model.evaluate(input_fn=eval_input_fn(features=X_test, labels=y_test, batch_size=len(X_test)))
```

```python
print("Test set accuracy: {accuracy:0.3f}".format(**eval_result))
```
